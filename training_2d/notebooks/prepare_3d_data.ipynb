```python
ashish_df_int = ashish_df.applymap(lambda x: int(x) if isinstance(x, bool) else x)   ## considered for all real + fake data
ashish_df_int_fake  = ashish_df_int[                                                     ## considered for all non real nodule data     
    ((ashish_df_int['nodule_fake'] == 1) & (ashish_df_int['nodule_real'] == 0)) |
    ((ashish_df_int['nodule_fake'] == 0) & (ashish_df_int['nodule_real'] == 0))
].reset_index(drop=True)
```

```python
ashish_df_int_fake.rename(columns = {"nodule_fake": 'nodule_fake_3d' ,"nodule_real": 'nodule_real_3d','regular' : 'regular_border', 'irregular' : 'irregular_border' },inplace = True)
ashish_df_int.rename(columns = {"nodule_fake": 'nodule_fake_3d' ,"nodule_real": 'nodule_real_3d','regular' : 'regular_border', 'irregular' : 'irregular_border' },inplace = True)
```

```python
final_columns = ['filename', 'nodule', 'regular_border', 'irregular_border', 'homogenous', 
                 'inhomogenous', 'calcified', 'intensity_1', 'intensity_2', 
                 'intensity_3', 'nodule_fake_3d', 'nodule_real_3d']
```

```python
ashish_df_int_fake = ashish_df_int_fake.drop(columns=['old_path', 'base_image'])[final_columns]
ashish_df_int = ashish_df_int.drop(columns=['old_path', 'base_image'])[final_columns]
```

```python
ashish_df_int_fake['real'] = 1 - ashish_df_int_fake['nodule_fake_3d']
ashish_df_int['real'] = 1 - ashish_df_int['nodule_fake_3d']
```

```python
ashish_df_int_fake['3d_data'] = 1 
ashish_df_int['3d_data'] = 1 
ashish_df_int_fake['artifact'] = 0 
ashish_df_int['artifact'] = 0
```

```python
ashish_df_int_fake['normal'] = 1 - ashish_df_int_fake['nodule']
ashish_df_int['normal'] = 1 - ashish_df_int['nodule']
```

```python
piyush_df_ = piyush_df.merge(old_data[['filename', 'calcified']], on='filename', how='left')
# piyush_df_['calcified'] = piyush_df_['calcified'].fillna(-100).astype(int)
piyush_df_['calcified'] = piyush_df_['calcified'].fillna(-100).astype(int)
```
