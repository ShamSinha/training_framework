```python
import pandas as pd

# Path to the original CSV file
csv_file_path = '/raid/cxr/qxr_ln_data/LN_test/combined_test_csv_wo_padchest_09-09-24.csv'

# Load the existing CSV
df = pd.read_csv(csv_file_path)

# Filename to start from
start_filename = 'ccae584af41e42ea9b871f0f7b7ae2e8'

# Find the index of the row where the filename appears
start_idx = df.index[df['filename'] == start_filename].tolist()

if start_idx:
    start_idx = start_idx[0]  # Get the first occurrence of the filename
    # Slice the DataFrame from the start index
    df_sliced = df.iloc[start_idx:]

    # Path to the new CSV file
    new_csv_file_path = '/raid/cxr/qxr_ln_data/LN_test/remaining_test.csv'

    # Save the sliced DataFrame to a new CSV file
    df_sliced.to_csv(new_csv_file_path, index=False)
    print(f"New CSV created starting from {start_filename} and saved at {new_csv_file_path}")
else:
    print(f"Filename {start_filename} not found in the CSV.")
```
